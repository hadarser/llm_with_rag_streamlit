## Download RAG data

This notebook is used to download the `gretelai/synthetic_text_to_sql` from HuggingFace

```python
import json

import pandas as pd
import requests
```

```python
url = "https://datasets-server.huggingface.co/rows?dataset=gretelai%2Fsynthetic_text_to_sql&config=default&split=train&offset={offset}&length=100"
```

```python
requests.get(url.format(offset=100)).json()
```

```python
data_pages = []
i = 0
while i < 50:  # 50 pages of 100 rows each
# for i in range(50):  # 50 pages of 100 rows each
    data_pages.append(requests.get(url.format(offset=i * 100)).json())
    if i % 10 == 0:
        print(f"Downloaded {i} pages")
    i += 1
print(f"Downloaded {len(data_pages)} pages - done")
```

```python
page = data_pages[0]
print("Keys:", page.keys())
print(f"{page['partial']=}")
print(f"{page['num_rows_per_page']=}")
print("Features:")
print(json.dumps(page["features"], indent=2))
print("Row example:")
print(page["rows"][0].keys())
print(page["rows"][0]["row_idx"])
print(page["rows"][0]["row"])
print(page["rows"][0]["truncated_cells"])
```

```python
all_rows = []
for page in data_pages:
    try:
        all_rows.extend([r["row"] for r in page["rows"]])
    except TypeError:
        print("Error in page")
        break
full_df = pd.DataFrame(all_rows)
full_df
```

```python
assert len(full_df) == 5000
assert full_df.id.is_monotonic_increasing
assert full_df.id.is_unique
```

```python
full_df.to_csv("synthetic_text_to_sql.csv", index=False)
```
